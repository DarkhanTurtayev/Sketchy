```python
import os
import cv2
```

```python
def reduce_image_quality(input_path, output_path, scale_factor=1, quality=50):

    image = cv2.imread(input_path)
    if image is None:
        print(f"Не удалось загрузить изображение {input_path}. Пропуск...")
        return
    
    new_width = int(image.shape[1] * scale_factor)
    new_height = int(image.shape[0] * scale_factor)
    
    resized_image = cv2.resize(image, (new_width, new_height), interpolation=cv2.INTER_CUBIC)
    
    cv2.imwrite(output_path, resized_image, [int(cv2.IMWRITE_JPEG_QUALITY), quality])
    print(f"Обработано и сохранено {output_path}")

def process_images(input_dir, output_dir, scale_factor=1, quality=50):
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    for filename in os.listdir(input_dir):
        if filename.endswith(".jpg") or filename.endswith(".png"):
            input_path = os.path.join(input_dir, filename)
            output_path = os.path.join(output_dir, filename)
            reduce_image_quality(input_path, output_path, scale_factor, quality)
```

```python
input_dir = '/Users/macstudiod/Desktop/Upscaler/output'  # Укажите путь к директории с исходными изображениями
output_dir = '/Users/macstudiod/Desktop/Upscaler/LR'  # Укажите путь к директории для сохранения обработанных изображений

process_images(input_dir, output_dir, scale_factor=0.5, quality=50)
```
